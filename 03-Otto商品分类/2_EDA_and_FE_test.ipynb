```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dpath = './data/'

test = pd.read_csv(dpath + 'Otto_test.csv')
test.head()
```

## 特征工程
测试数据的特征工程和训练数据一样 特征编码的模型由训练集得到，此次将模型装载进来就好

```python
test_id = test['id']
X_test = test.drop(['id'], axis=1)
col_names_org = X_test.columns
```

```python
# 1. feat编码：log(x+1)
X_test_log = np.log1p(X_test)

col_names_log = col_names_org + '_log'

X_test_log = pd.DataFrame(columns=col_names_log, data=X_test_log.values)
X_test_log.head()
```

```python
# 2. feat编码：TF-IDF
import pickle

tfidf = pickle.load(open('tfidf.pkl', 'rb'))

X_test_tfidf = tfidf.transform(X_test).toarray()

col_names_tfidf = col_names_org + '_tfidf'

X_test_tfidf = pd.DataFrame(columns=col_names_tfidf, data=X_test_tfidf)
X_test_tfidf.head()
```

```python
# 数据归一化
mms_org = pickle.load(open("MinMaxScaler_org.pkl", 'rb'))
mms_log = pickle.load(open('MinMaxScaler_log.pkl', 'rb'))
mms_tfidf = pickle.load(open('MinMaxScaler_tfidf.pkl', 'rb'))

feat_names_org = X_test.columns
feat_names_log = X_test_log.columns
feat_names_tfidf = X_test_tfidf.columns

X_test = mms_org.transform(X_test)
X_test_log = mms_log.transform(X_test_log)
X_test_tfidf = mms_tfidf.transform(X_test_tfidf)
```

```python
test_org = pd.concat([test_id, pd.DataFrame(columns=feat_names_org, data=X_test)], axis=1)
test_org.to_csv(dpath + 'Otto_FE_test_org.csv', index=False, header=True)

test_log = pd.concat([test_id, pd.DataFrame(columns=feat_names_log, data=X_test_log)], axis=1)
test_log.to_csv(dpath + 'Otto_FE_test_log.csv', index=False, header=True)

test_tfidf = pd.concat([test_id, pd.DataFrame(columns=feat_names_tfidf, data=X_test_tfidf)], axis=1)
test_tfidf.to_csv(dpath + 'Otto_FE_test_tfidf.csv', index=False, header=True)
```
